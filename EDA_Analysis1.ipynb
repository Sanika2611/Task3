```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds=pd.read_csv("employee_dataset.csv")
ds
```

```python
ds.info()
```

#categorical variable visualization

```python
#employee count per department

sns.countplot(data=ds,x='Department',palette='Set2')
plt.title("Employee count per department")
```

Finance department having more employee count followed by engineering and marketing

```python
sns.countplot(data=ds,x='Gender',palette='coolwarm')
plt.title("Gender Distribution")
plt.show()
```

Female entries are more in count as compared to male count

```python
#Employee Location Distribution

sns.countplot(data=ds,x='Location',palette='Spectral')
plt.title("NUmber of employees by location")
#plt.show()
plt.xticks(rotation=60)
```

```python
#Age Distribution

sns.histplot(ds['Age'],bins=10,kde=True)
plt.title("Age Distribution")
```

Age 45-50 has more employees group as per KDE plot,then age group 25-30 

```python
#Years of experience Distribution

sns.histplot(ds["Years_of_Experience"],bins=15,kde=True)
plt.title("Years of experience")
plt.show()
```

From kde plot,we say that there are more employees who has experience of 2.5 to 5 years.

```python
#Monthly salary distribution

sns.boxplot(data=ds,y='Monthly_Salary',color='lightpink')
plt.title("Salary Speed")
```

min=20,000 max=1,20,000 average=75000

```python
#Average salary by department

sns.barplot(data=ds,x='Department',y='Monthly_Salary',estimator=np.mean,palette="viridis")
plt.title("Average salary by department")
plt.xticks(rotation=45)
```

Finance department have more average salary as compared to other departments..

```python
#Performance score by department

sns.boxplot(data=ds,x='Department',y='Performance_Score',palette='pastel')
plt.title("Performance score by department")
```

Marketing department performance score is more than other department

```python
#Salary distribution by gender


sns.violinplot(data=ds,x='Gender',y='Monthly_Salary',palette='muted')
plt.title("Salary by gender")
```

Male has more salary range between 1,00,000 to 1,20,000 as compare to Female.

```python
sns.heatmap(ds[['Age','Years_of_Experience','Monthly_Salary','Performance_Score']].corr(),annot=True,cmap='coolwarm')
plt.title("Correlation Matrix")
```

```python
#Salary vs Experience Scatter Plot

sns.scatterplot(data=ds,x='Years_of_Experience',y='Monthly_Salary',hue='Department')
plt.title("Salary vs.Experience by department")
```
